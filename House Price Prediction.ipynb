# Importing libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

# Loading Dataset

```python
data = pd.read_csv("USA_housing.csv")
data.head(10)
```

# Dropping not import attribute i.e Address(since it is in string form)

```python
data = data.drop(['Address'], axis=1)
```

```python
data.head()
```

#  Checking missing data by plotting heatmap

```python
sns.heatmap(data.isnull())
```

no missing value from the above graph

# Train Test split data

```python
X = data.drop("Price", axis=1)
Y = data['Price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

# Training and predicting

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

prediction

```python
prediction = model.predict(X_test)
```

```python
prediction
```

# Evaluation

```python
#using root mean square error
error = np.sqrt(metrics.mean_absolute_error(Y_test, prediction))
```
